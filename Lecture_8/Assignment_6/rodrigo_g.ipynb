```python
def uni_geo( row ):
    
    query = f"{ row[ 'UNIVERSIDAD' ] }, { row[ 'DEPARTAMENTO' ] }, { row[ 'PROVINCIA' ] }"
    
    gmaps = googlemaps.Client( key = key )
    result_api = gmaps.geocode( query , region = 'PE' )
    
    try:
        lat = result_api[0]['geometry']['location']['lat']
        lon = result_api[0]['geometry']['location']['lng']   
    except:
        lat = np.nan
        lon = np.nan
    
    return  ( lat, lon )
    
    
```

```python
endpoint = 'https://maps.googleapis.com/maps/api/directions/json?'

models = [ 'best_guess', 'pessimistic', 'optimistic' ]

departure_time = 'now'

mode = 'driving'

region = 'pe'

travel_times = { model: [] for model in models }

distances = { model: [] for model in models }

```

```python
base_5p = base.sample( 5, random_state = 1 ).reset_index( drop = True)
```

```python
base_5p
```

```python
travel_times = { model: [] for model in models }

distances = { model: [] for model in models }

for index, row in tqdm( base_5p.iloc[ :, -4:].iterrows() ):
    
    origin = ','.join( [ str( row[ 0 ] ), str( row[ 1 ] ) ] )
    
    destination = ','.join( [ str( row[ 2 ] ), str( row[ 3 ] ) ] )
    
    for model in models:
        
        traffic_model = model
        
        nav_request = 'origin={}&destination={}&departure_time={}&traffic_model={}&mode={}&region={}&key={}'.format(origin, destination, departure_time, traffic_model, mode, region, key)
        
        request = endpoint + nav_request
        
        try:
            response = urllib.request.urlopen( request ).read()
            
            directions = json.loads( response )
            
            legs = directions[ 'routes' ][ 0 ][ 'legs' ][ 0 ]
            
            travel_times[ model ] += [ legs[ 'duration' ][ 'text' ] ]
            
            distances[ model ] += [ legs[ 'distance' ][ 'text' ] ]
        
        except Exception as e:
            
            travel_times[ model ] += "nan"
            distances[ model ] += "nan"
    
```

```python
for model in models:
    
    base_5p[ f"travel_time_{ model }" ] = travel_times[ model ]
    base_5p[ f"distance_{ model }" ] = distances[ model ]
```
