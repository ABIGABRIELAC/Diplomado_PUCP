```python
def uni_geo( row ):
    
    query = f"{ row[ 'UNIVERSIDAD' ] }, { row[ 'DEPARTAMENTO' ] }, { row[ 'PROVINCIA' ] }"
    
    gmaps = googlemaps.Client( key = key )
    result_api = gmaps.geocode( query , region = 'PE' )
    
    try:
        lat = result_api[0]['geometry']['location']['lat']
        lon = result_api[0]['geometry']['location']['lng']   
    except:
        lat = np.nan
        lon = np.nan
    
    return  ( lat, lon )
    
    
```

```python
uni_geo( univs.iloc[ -2, : ] )
```

```python
univs[ 'COORDS' ] = univs.apply( lambda row: uni_geo( row ), axis = 1 )
```

```python
univs[ 'Latitude_univ' ] = univs[ 'COORDS' ].apply( lambda x: [ y for y in x ][ 0 ] )
univs[ 'Longitude_univ' ] = univs[ 'COORDS' ].apply( lambda x: [ y for y in x ][ 1 ] )
```

```python
univs = univs.drop( ['COORDS' ], axis = 1 )
```

```python
univs[ 'DEPARTAMENTO' ] = univs[ 'DEPARTAMENTO' ].apply( lambda x: unidecode.unidecode( x.upper() ) )
```

```python
univs.to_csv( r'_data/univs.csv', index = False, encoding = 'UTF-8-SIG' )
```
