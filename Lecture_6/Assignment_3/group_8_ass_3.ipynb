```python
def reg_ols( x : pd.DataFrame, y : pd.Series, intercept) -> pd.DataFrame:
    if not isinstance( x , pd.DataFrame ):
        raise TypeError( "X variable is not .DataFrame.")
        
    if not isinstance( y, pd.Series ):
        raise TypeError( "Y variable is not pd.Series.")
        
    base_datos = x
    base_datos['log(Cost)'] = y.tolist()
    
    if intercept == True:
        regresion = smf.ols( 'base_datos.iloc[:, 4] ~ base_datos.iloc[:, 0]+base_datos.iloc[:, 1]+base_datos.iloc[:, 2]+base_datos.iloc[:, 3]', data = base_datos)
        result = regresion.fit().summary2().tables[1].iloc[ :, [0]]
    
    else:
        regresion = smf.ols( 'base_datos.iloc[:, 4] ~ base_datos.iloc[:, 0]+base_datos.iloc[:, 1]+base_datos.iloc[:, 2]+base_datos.iloc[:, 3]-1', data = base_datos)
        result = regresion.fit().summary2().tables[1].iloc[ :, [0]]
        
    return result
```
