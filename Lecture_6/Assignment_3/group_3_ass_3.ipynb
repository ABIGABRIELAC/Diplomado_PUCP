```python
def B_OLS(X:pd.DataFrame, y:pd.Series, intercept:bool=True) -> pd.DataFrame:
    
    if not isinstance(X, pd.DataFrame):
        raise TypeError('X is not DF')
        
    if not isinstance(y, pd.DataFrame):
        raise TypeError('y is not a pd S')
        
    if not isinstance(intercept, bool):
        raise TypeError('intercept is not B')
    
    if intercept:
        X_ = X.copy()
        X_.insert(0, 'Intercept', 1)
        betas = np.linalg.inv(X_.T @ X_) @ X_.T @ y
        betas['Params'] = X_.columns.tolist()
        betas.set_index('Params', inplace=True)
        betas.rename(columns={'lnct':'Coefs'}, inplace=True)
        return betas
    else:
        betas = np.linalg.inv(X.T @ X) @ X.T @ y
        betas['Params'] = X.columns.tolist()
        betas.set_index('Params', inplace=True)
        betas.rename(columns={'lnct':'Coefs'}, inplace=True)
        return betas
```

```python
def var_OLS( X : pd.DataFrame, y : pd.Series, intercept = True ):
        
    if not isinstance( X, pd.DataFrame ):
        raise TypeError( 'X is not DF' )
    
    if not isinstance( y, pd.Series ):
        raise TypeError( 'y is not a pd S' )
    
    if not isinstance( intercept, bool ):
        raise TypeError( 'intercept is not B' )
        
    if len( X ) != len( y ):
        raise ValueError( 'rows in x and y differ' )
    
    X_mat = X.copy()
    y_vec = y.copy()
    
    if intercept == True:
        const = np.ones( len( X_mat ) )
        X_mat.insert(0, "Intercept", const)
    
    var_names = list( X_mat.columns )
    X_mat = X_mat.to_numpy()
    y_vec = y_vec.to_numpy()
    
    xtx_inv = np.linalg.inv( np.dot( np.transpose( X_mat ), X_mat ) )
    coefs = np.dot( xtx_inv, np.dot( np.transpose( X_mat ), y_vec ) )
    
    errores=np.add(y_vec,-np.dot( X_mat , coefs ))
    var_error=np.dot(np.transpose(errores),errores)/(len(errores)-4-1)

    var_coef=var_error*xtx_inv
    
    var_coef = pd.DataFrame( var_coef, index = var_names, columns=var_names)

    
    return var_coef
```
