X (covariables) must be a pd.DataFrame and y (endog) must be a pd.Series and intercept can be True or False, by default True. The output of the function should look like var_OLS_output where the columns and the index use the name of X columns. Apply your function to find $\mathbb{V}{\rm ar} (\widehat{\boldsymbol{\beta}}^{(OLS)})$ of the equation below. Use greene data. Hint: Use numpy, def function, and columns method, link.


```python
2.
file_to_open = open( r'../../_data/var_OLS_output.pkl' , "rb" )
var_OLS_output = pickle.load( file_to_open )
var_OLS_output

def var_OLS (X:pd.DataFrame, y:pd.Series, intercept:bool = True ) ->pd.DataFrame:
    if not isinstance(X, pd.DataFrame):
        raise TypeError('X is not a DataFrame')
    
    if not isinstance(y, pd.Series):
        raise TypeError('y is not a Series')
    
    if not isinstance(intercept, bool):
        raise TypeError('intercept is not a boolean')
        
    X_n = X.copy() 
    x_names = list(X_n.columns)
        
        
    if intercept == True:

        x1 = np.hstack(( np.ones(len(X_n)).reshape(-1, 1 ) , X_n))
        b = np.linalg.inv(x1.T @ x1) @ x1.T @ y
        #y predicted
        y_hat = np.dot(x1,b)
        
        #residual
        e = y - y_hat
        
        SCR = np.sum((e**2)/(len(y)-len(b)))
        var_b = SCR * np.linalg.inv(x1.T @ x1)
    
    

        x_names.insert(0,"Intercept")
        
        var_b = pd.DataFrame(var_b, columns = x_names)
        
        var_b[''] = x_names
        var_b.set_index('', inplace=True)
        
    else:
        b_2 = np.linalg.inv(X_n.T @ X_n) @ X_n.T @ y
        
        #y predicted
        y_hat2 = np.dot(X_n ,b)
        e_2 = y - y_hat2
        
        SCR_2 = np.sum((e_2**2)/(len(y)-len(b_2)))

        var_b = SCR_2 * np.linalg.inv(X_n.T @ X_n)
        

        var_b = pd.DataFrame(var_b, columns = x_names)
        
        var_b[''] = x_names
        var_b.set_index('', inplace=True)
        
    return var_b
```
