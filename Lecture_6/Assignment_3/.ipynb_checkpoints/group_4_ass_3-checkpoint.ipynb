$$
\begin{aligned}

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
greene = pd.read_csv(r"..\..\_data/christensen_greene_f4.csv")
greene
```

```python
greene.columns = [col.strip() for col in greene.columns]
greene.columns
```

```python
results = smf.ols( 'np.log( COST ) ~ np.log( Q ) + np.log(PL)+ np.log(PF) + np.log(PK)', data = greene)
```

```python
X_input = pd.DataFrame(results.exog, columns = results.exog_names).iloc[ : , 1: ]
y_input = pd.Series(results.endog , name = results.endog_names )
intercept_input = True
```

```python
beta_OLS_output = results.fit().summary2().tables[1].iloc[ :, [0]]
beta_OLS_output
```

```python
only_cost = greene.COST.values # para y_input
only_Q = greene.Q.values # para lnQ
only_PL = greene.PL.values # para lnPL
only_PK = greene.PK.values # para lnKP
only_PF = greene.PF.values # para lnPF
```

```python
# Hallar y_input
y_input = pd.DataFrame ([np.log( only_cost )])
y_input = y_input.T.iloc[:,0]
y_input.rename("np.log(COST)", inplace =True)
print ( y_input )
```

```python
# Hallar X_input
x_input_q = np.log( only_Q )
x_input_pl = np.log( only_PL )
x_input_pk = np.log( only_PK )
x_input_pf = np.log( only_PF )
x_input = pd.DataFrame([ x_input_q, x_input_pl, x_input_pk, x_input_pf ]).T
x_input.rename(columns = {0:'np.log(Q)', 1:'np.log(PL)',
                    2:'np.log(PF)', 3:'np.log(PK)'}, inplace = True)
print (x_input )
```

```python
#Model
lnct = np.log(only_cost)
lnq = np.log(only_Q)
lnp1 = np.log(only_PL)
lnp2 = np.log(only_PK)
lnp3 = np.log(only_PF)
lnq_2 = np.multiply(lnq,lnq)
lnqlnp1 = np.multiply(lnq,lnp1)
lnqlnp2 = np.multiply(lnq,lnp2)
lnqlnp3 = np.multiply(lnq,lnp3)
lnp1_2 = np.multiply(lnp1,lnp1) 
lnp2_2 = np.multiply(lnp2,lnp2) 
lnp3_2 = np.multiply(lnp3,lnp3) 
lnp1lnp2 = np.multiply (lnp1,lnp2)
lnp1lnp3 = np.multiply (lnp1,lnp3)
lnp2lnp3 = np.multiply (lnp2,lnp3)
```

```python
results_r = smf.ols( 'lnct ~ lnq + lnq_2 + lnqlnp1 + lnqlnp2 + lnqlnp3 + lnp1 + lnp2 + lnp3 + lnp1_2 + lnp2_2 + lnp3_2 + lnp1lnp2 + lnp1lnp3 + lnp2lnp3', 
        data = greene)
results_r
```

```python
X_r = pd.DataFrame(results_r.exog, columns = results_r.exog_names).iloc[ : , 1: ]
y_r = pd.Series(results_r.endog , name = results_r.endog_names )
intercept_r = True
beta_OLS_output_r = results_r.fit().summary2().tables[1].iloc[ :, [0]]
beta_OLS_output_r
```
