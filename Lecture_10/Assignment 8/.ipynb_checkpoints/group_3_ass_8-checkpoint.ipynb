#### 6. Use geopandas to plot the number of cases by the department for all the years using subplots. Every subplot for each year. Do not forget to indicate the color of NA values. Use this shapefile. For this task, you will have to aggregate shapefiles at the department level.

```python
dpt_shapes = gpd.read_file(r'../../_data/INEI_LIMITE_DEPARTAMENTAL/INEI_LIMITE_DEPARTAMENTAL.shp')
```

```python
dengue_dpt_all = dengue.groupby(['Departamento','Año'], as_index = False )[ [ 'Casos' ] ].sum()
```

```python
departments = dengue_dpt_all['Departamento'].unique().tolist()
departments = set(departments)
d = dpt_shapes['NOMBDEP'].unique().tolist()
d = set(d)
missing = list(sorted(d - departments))
years = dengue_dpt_all['Año'].unique().tolist()
```

```python
rows = []
for miss_dpt in missing:
    for year in years:
        rows.append([miss_dpt, year, np.nan])
```

```python
missing_departments = pd.DataFrame(rows, columns=['Departamento', 'Año', 'Casos'])
dengue_dpt_all = dengue_dpt_all.append(missing_departments)
dengue_dpt_all = dengue_dpt_all.sort_values('Departamento', ascending=True)
dengue_dpt_all
```

```python
dpt_map = dpt_shapes.merge(dengue_dpt_all, left_on='NOMBDEP', right_on='Departamento', how='left')
```

```python
years = dengue_dpt_all['Año'].unique().tolist()
fig, axes = plt.subplots(3, 3, figsize=(50,50))

for ax, year in zip(axes.ravel(), years):
    cmap = plt.cm.Purples
    dpt_map[dpt_map['Año'] == year].plot( figsize = (50,50), column = 'Casos', cmap = cmap, linestyle = 'dotted', edgecolor = 'black', legend = True, missing_kwds = dict( color = '#D0D0D0' ), ax=ax )
    ax.set_title('Casos de Dengue Departamental - {}'.format(year))

axes.flat[7].set_visible(False)
axes.flat[8].set_visible(False)
```
