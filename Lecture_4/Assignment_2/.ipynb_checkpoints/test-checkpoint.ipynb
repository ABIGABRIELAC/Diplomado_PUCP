```python
years = ['2015', '2016', '2017','2018' ,'2019']
all_data = {}
columns = []
```

```python
for year in years:
    
    dict_ = {}
    
    path1 = r"C:\Users\Usuario\Desktop\R\Curso\Diplomado_PUCP\_data\endes\{}\REC0111.SAV".format(year)
    path2 = r"C:\Users\Usuario\Desktop\R\Curso\Diplomado_PUCP\_data\endes\{}\RE223132.SAV".format(year)
    path3 = r"C:\Users\Usuario\Desktop\R\Curso\Diplomado_PUCP\_data\endes\{}\RE516171.SAV".format(year)
    
    rec_1 = pd.read_spss(path1)
    rec_2 = pd.read_spss(path2)
    rec_3 = pd.read_spss(path3)
    
    with sav.SavHeaderReader(path1, ioUtf8=True) as header:
        var_labels1 = header.all()
        value_labels1 = var_labels1.valueLabels

    with sav.SavHeaderReader(path2, ioUtf8=True) as header:
        var_labels2 = header.all()
        value_labels2 = var_labels2.valueLabels

    with sav.SavHeaderReader(path3, ioUtf8=True) as header:
        var_labels3 = header.all()
        value_labels3 = var_labels3.valueLabels
    
    var_labels1 = var_labels1.varLabels
    var_labels2 = var_labels2.varLabels
    var_labels3 = var_labels3.varLabels
    
    cols2 = rec_2.columns.tolist()
    index_dict = {}
    
    for idx, value in enumerate(cols2):
        index_dict[idx] = value
      
    if year == '2018':  ## Year 2018 doesn't have the variable 'V007'
        rec_1_1 = rec_1[["CASEID","V000","V001","V002","V003","V004","V008","V009","V010","V011","V012","V024","V102", "V120","V121","V122","V123","V124","V125","V127","V133"]]
        rec_2_1 = rec_2[["CASEID","V201","V218","V301","V302","V323","V323A","V325A","V326","V327","V337","V359","V360","V361", "V362","V363","V364","V367","V372","V372A","V375A","V376","V376A","V379","V380"]]
        rec_3_1 = rec_3[["CASEID","V501","V502","V503","V504","V505","V506","V507","V508","V509","V510","V511","V512","V513", "V525","V613","V714","V715"]]
    
    else:
        rec_1_1 = rec_1[["CASEID","V000","V001","V002","V003","V004","V007","V008","V009","V010","V011","V012","V024","V102", "V120","V121","V122","V123","V124","V125","V127","V133"]]
        rec_2_1 = rec_2[["CASEID","V201","V218","V301","V302","V323","V323A","V325A","V326","V327","V337","V359","V360","V361", "V362","V363","V364","V367","V372","V372A","V375A","V376","V376A","V379","V380"]]
        rec_3_1 = rec_3[["CASEID","V501","V502","V503","V504","V505","V506","V507","V508","V509","V510","V511","V512","V513", "V525","V613","V714","V715"]]
    
    
    
    new_var_labels1 =  { key: var_labels1[key] for key in rec_1_1.columns.tolist() }
    new_var_labels2 =  { key: var_labels2[key] for key in rec_2_1.columns.tolist() }
    new_var_labels3 =  { key: var_labels3[key] for key in rec_3_1.columns.tolist() }
    
    new_value_labels1 = {}
    
    for key in rec_1_1.columns.tolist():
        try:
            new_value_labels1[key] = value_labels1[key]
        except:
            continue
    
    new_value_labels2 = {}
    
    for key in rec_2_1.columns.tolist():
        try:
            new_value_labels2[key] = value_labels2[key]
        except:
            continue
        
    new_value_labels3 = {}
    
    for key in rec_3_1.columns.tolist():
        try:
            new_value_labels3[key] = value_labels3[key]
        except:
            continue
        
    rec_1_1['year'] = year
    new_var_labels1['year'] = 'Year of the survey'
    
    endes = rec_1_1.merge(rec_2_1, on='CASEID', how='left')
    endes = endes.merge(rec_3_1, on='CASEID', how='left')
    
    var_labels = new_var_labels1
    var_labels.update(new_var_labels2)
    var_labels.update(new_var_labels3)
    
    value_labels = new_value_labels1
    value_labels.update(new_value_labels2)
    value_labels.update(new_value_labels3)
    
    dict_['data'] = endes
    dict_['var_labels'] = var_labels
    dict_['value_labels'] = value_labels
    
    all_data['year_{}'.format(year)] = dict_
    columns = endes.columns.tolist() #Export columns to create the dataframe that will store the concatenated 2015-19 df
```

```python
endes_data_2015_2019 = pd.DataFrame(columns=columns)

for year in years:
    df = all_data['year_{}'.format(year)]['data']
    endes_data_2015_2019 = pd.concat([endes_data_2015_2019,df], axis=0)

endes_data_2015_2019 = endes_data_2015_2019.reset_index()
```
