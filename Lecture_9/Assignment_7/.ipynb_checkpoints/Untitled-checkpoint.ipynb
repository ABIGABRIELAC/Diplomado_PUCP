#### 1. Data establecimientos de salud cajamarca

```python
estb = pd.read_excel(r'../../_data/cajamarca_data_helth_estb.xlsx')
estb.shape
```

```python
estb.NOMBDEP[estb.NOMBDEP != "CAJAMARCA"]
```

```python
estb_ca = estb[estb.NOMBDEP == 'CAJAMARCA']
estb_ca.shape
```

#### 2. Data centroides distritales

```python
centroids = pd.read_excel(r'../../_data/peru_districts_centroids.xlsx')
cen_ca = centroids[centroids.NOMBDEP == 'CAJAMARCA']
cen_ca.NOMBDEP[cen_ca.NOMBDEP != 'CAJAMARCA']
```

```python
cen_ca['Centroid_Longitude'].isnull().sum()
```

#### 3. Agregar columnas centroides distritales en base de eess

```python
data_ca = pd.merge( estb_ca, cen_ca, how = 'left', on = ['CCDD',\
       'NOMBDEP', 'CCPP', 'NOMBPROV', 'CCDI', 'NOMBDIST', 'CAPITAL', 'UBIGEO',\
       'IDPROV', 'CODIGO', ] )
```

```python
print(data_ca.latitud.isnull().sum())
print(data_ca.longitud.isnull().sum())
```

```python
# Crear columna distancia euclidiana
data_ca['eu_dist'] = ""
```

```python
# llenar columna distancia euclidiana a parti de f√≥rmula para de cada fila
i = 0
for array in range(len(data_ca.index)):
    estb_array = np.array(data_ca[['longitud', 'latitud']])
    cen_array = np.array(data_ca[['Centroid_Longitude', 'Centroid_Latitude' ]])
    data_ca.iloc[i,-1] = sqrt( (estb_array[i,0]-cen_array[i,0])**2 + (estb_array[i,1]-cen_array[i,1])**2 )
    
    i = i + 1
```

```python
data_ca.tail()
```

```python
# verificar
j = -2
eu_dist = sqrt( (estb_array[j,0]-cen_array[j,0])**2 + (estb_array[j,1]-cen_array[j,1])**2 )
eu_dist
```

```python
# guardamos df con columna euclidiana en data_dist
data_dist = data_ca.copy()
data_dist.iloc[:,[0,14,16,25]]
```

#### 4. Find the closest health establishment to Cajamarca district centroids

```python
# para cada distrito de cada provincia, ¬øcual es la distancia m√≠nima entre el centroide distrital y el establecimiento de salud m√°s cercano?
data_dist.groupby(['NOMBPROV','NOMBDIST'], as_index = False).eu_dist.min()
# aplicar min() despu√©s del group by() a todas las variales distorsiona el eess que presenta la menor distancia respecto del centroide distrital: (como variable id_eess es n√∫merica se toma el n√∫mero de id_eess m√°s peque√±o)
```

```python
%matplotlib inline
```

```python
data_dist.groupby(['NOMBPROV','NOMBDIST'], as_index = True).eu_dist.min().plot(kind = 'bar')
```

```python
# eliminamos variables redundantes en data_dist
data_dist.drop(['ind', 'IDPROV', 'CODIGO', 'CNT_CCPP' ], axis = 1, inplace = True)
```

```python
# ordenamos los valores seg√∫n distancia euclidiana, provincia y  distrito, en ese orden y guardamos en 'dist_min_dist'
data_min_dist = data_dist.sort_values(['NOMBDIST','eu_dist','NOMBPROV' ], ascending = True)
```

```python
# Nos quedamos solo con valores m√°s peque√±os de 'eu_dist' para cada distrito de cada provincia
data_min_dist = data_min_dist.drop_duplicates(subset = ['NOMBPROV','NOMBDIST' ], keep = 'first')
data_min_dist[data_min_dist['NOMBPROV'] == 'SAN MARCOS']
```

```python
data_min_dist.NOMBDIST[data_min_dist.NOMBPROV == 'SAN MARCOS' ]
```

```python
data_min_dist.NOMBDIST.nunique()
```

```python
#guardamos data con el 'eess' cuya distancia al centroide distrittal sean menor para las 13 provincias de Cajamarca
data_min_dist.to_csv(r'data_min_dist.csv',  index = False, encoding = 'UTF-8-SIG' )
```
