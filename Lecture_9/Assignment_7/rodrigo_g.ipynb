```python
HE = pd.merge( HE_caj, cen_caj, on = [ 'NOMBPROV', 'NOMBDIST' ], validate = 'm:1' )
```

```python
HE
```

```python
def distance( a, b ):
    dst = np.sqrt( ( a[ 0 ] + b[ 0 ] ) ^ 2 + ( a[ 1 ] + b[ 1 ] ) ^ 2 )
    return dst
```

```python
HE[ 'DISTANCE' ] = HE.apply( lambda row: np.sqrt( ( row[ 'Centroid_Latitude' ] - row[ 'latitud' ] ) ** 2 + ( row[ 'longitud' ] - row[ 'Centroid_Longitude' ] ) ** 2 ), axis = 1 )
```

```python
HE_near = HE.sort_values( [ 'DISTANCE' ] ).groupby( [ 'NOMBPROV', 'NOMBDIST' ] ).head( 1 )
```
